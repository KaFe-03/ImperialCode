```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
D, Phase, std = np.loadtxt("D:\_IMPERIAL\Laboratory\Speed of Sound/Labs 2 Speed of sound.csv", skiprows=2, delimiter=",", unpack=True)
```

```python
fit_sos,cov_sos = np.polyfit(D,Phase,1,w=1/std,cov=True)
print(fit_sos)
print(cov_sos)
```

```python
sig_0 = np.sqrt(cov_sos[0,0])
print('Slope = %.3e +/- %.3e' %(fit_sos[0],sig_0))
```

```python
psos=np.poly1d(fit_sos)
print("Equation of Line")
print(psos)
params = {
    "axes.labelsize": 12,
    "font.size": 12,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
    "figure.figsize": [12,8]
}
plt.rcParams.update(params)
plt.title("Speed of Sound Data")
plt.ylabel("Phase (Radian)")
plt.xlabel("Distance (m)")
plt.grid(axis="both")
plt.plot(D,Phase, "x", mew=1, ms=8, color="red")
#plt.errorbar(D,Phase,yerr=std, xerr=0.00001, fmt='o', mew=2, ms=3, capsize=4)
plt.plot(D,psos(D))
plt.show()
```
